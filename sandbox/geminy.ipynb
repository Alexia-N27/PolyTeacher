```python
import pandas as pd
import re
```

```python
# Fonction pour nettoyer le texte
def clean_text(text):
    # Supprime les caractères spéciaux comme l'apostrophe
    return re.sub(r"[^\w\s]", "", text)
```

```python
data = {
    'Français': ["Bonjour", "Merci", "Comment tu t'appelles ?", "Où sont les toilettes ?", "Que fais tu dans la vie ?"],
    'Spanish': [None, None, None, None, None],
    'Italien': [None, None, None, None, None],
    'English': [None, None, None, None, None]
}
```

```python
dataframe = pd.DataFrame(data)
dataframe['Français'] = dataframe['Français'].apply(clean_text)
dataframe
```

```python
import google.generativeai as genai

# def translator(prompt, language):
#     api_key = ""

#     prompt = f"""
#     Traduis "{prompt}" en "{language}". La réponse ne doit contenir que la traduction.
#     """

#     genai.configure(api_key=api_key)

#     model = genai.GenerativeModel("gemini-1.5-flash")
#     response = model.generate_content(prompt)

#     return response.text.strip()

def translator(prompt, language):
    try:
        api_key = ""

        prompt = f"""
        Traduis "{prompt}" en "{language}". La réponse ne doit contenir que la traduction.
        """

        genai.configure(api_key=api_key)

        model = genai.GenerativeModel("gemini-1.5-flash")
        response = model.generate_content(prompt)

        # Afficher le contenu pour debug
        print(f"Traduction pour '{prompt}': {response.text.strip()}")

        return response.text.strip()
    except Exception as e:
        print(f"Erreur lors de la traduction de '{prompt}' en {language}: {e}")
        return None  # Retourner None en cas d'erreur
```

```python
dataframe['Spanish'] = dataframe['Français'].apply(lambda x: translator(x, "Espagnol"))
dataframe['Italien'] = dataframe['Français'].apply(lambda x: translator(x, "Italien"))
dataframe['English'] = dataframe['Français'].apply(lambda x: translator(x, "Anglais"))
```
